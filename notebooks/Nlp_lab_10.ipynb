# Lab 10: VADER Sentiment Analysis
Evaluate ten example sentences with NLTK's VADER sentiment analyzer.

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

# Download the lexicon once (cached after first run)
nltk.download('vader_lexicon')

sia = SentimentIntensityAnalyzer()
```

```python
# Ten labelled sentences to score
sentences = [
    ("I absolutely loved the new design; it's fantastic!", "pos"),
    ("The movie was a complete waste of time.", "neg"),
    ("I'm not sure how I feel about this product.", "neu"),
    ("This service is quick, friendly, and reliable.", "pos"),
    ("The meal was cold and tasteless.", "neg"),
    ("What an outstanding performance by the team today!", "pos"),
    ("The update didn't fix any of the issues I had.", "neg"),
    ("Weather is okay, nothing special today.", "neu"),
    ("Customer support resolved my issue in minutes.", "pos"),
    ("The directions were confusing and poorly written.", "neg"),
]
```

```python
def label(compound: float) -> str:
    if compound >= 0.05:
        return "pos"
    if compound <= -0.05:
        return "neg"
    return "neu"

rows = []
correct = 0
for text, true_label in sentences:
    scores = sia.polarity_scores(text)
    pred = label(scores["compound"])
    correct += pred == true_label
    rows.append({
        "sentence": text,
        "compound": round(scores["compound"], 3),
        "pos": scores["pos"],
        "neu": scores["neu"],
        "neg": scores["neg"],
        "pred": pred,
        "true": true_label,
        "match": pred == true_label,
    })

accuracy = correct / len(sentences)
print(f"Accuracy on the 10-sentence sample: {accuracy:.2%}")
for row in rows:
    print(f"\n{row['sentence']}")
    print(
        f"  compound={row['compound']} pos={row['pos']:.2f} neu={row['neu']:.2f} neg={row['neg']:.2f}"
    )
    print(f"  true={row['true']} predicted={row['pred']} match={row['match']}")
```
