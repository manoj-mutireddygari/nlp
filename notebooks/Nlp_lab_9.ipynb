```python
from nltk.tokenize import word_tokenize
from nltk import pos_tag
import nltk

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger_eng')

text = "the has big happiness cats see running seen on you."
tokens = word_tokenize(text)
tagged_words = pos_tag(tokens)

print(tagged_words)
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
text = "the has big happiness cats see running seen on you."
doc = nlp(text)
for token in doc:
    print(f"{token.text}:{token.pos_}")
```

```python
import pandas as pd
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('punkt_tab')
text= "when i was in bangalore, i used to have free lunches from company"
from nltk.stem import WordNetLemmatizer
token=nltk.word_tokenize(text)
lemmatizer=WordNetLemmatizer()
result=[]
for token in token:
    lemma_token=lemmatizer.lemmatize(token)
    result.append({"Original": token,"NLTK Lemmatizer":lemma_token,"same": token==lemma_token})
print(result)
results=pd.DataFrame(result)
print(results)
```

```python
import spacy
nlp=spacy.load('en_core_web_sm')
text="when i was in bangalore, i used to have free lunches from company"
doc=nlp(text)
result=[]
for token in doc:
    result.append({"original":str(token),"spacylemmatizer":token.lemma_,"same":str(token==token.lemma_)})
print(result)
results=pd.DataFrame(result)
print(results)
```
