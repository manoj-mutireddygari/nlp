```python
import numpy as np
import pandas
import nltk
import string
```

Character Tokenization

```python
a = "I am Fine"
b = list(a)
print(b)
```

Word Tokenization

```python
text = "I am learning NLP from scratch!"
text = text.lower()
print(text)

words = text.split()
print(words)
```

```python
new_words = []
for word in words:
  new_words.append(word.strip(string.punctuation))
print(new_words)
```

OR another Method for Word Tokenization

```python
from nltk.tokenize import word_tokenize
nltk.download('punkt_tab')
text2 = 'I am Fine'
wt = word_tokenize(text2)
print(wt)
```

N-Gram Tokenization

```python
from nltk.util import ngrams
text1 = "How are"
c = ngrams(text1,4)
for gram in c:
  print(gram)
```

Sentence Tokenizer

```python
from nltk.tokenize import sent_tokenize
nltk.download('punkt_tab')
text2 = 'I am Fine. How are You?'
wt = sent_tokenize(text2)
print(wt)
```
